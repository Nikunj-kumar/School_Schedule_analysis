```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#Selecting all data from the website
website = 'https://ischool.syr.edu/classes/?page'
w_data = pd.DataFrame()
for i in range(1,3):
    data = pd.read_html(website+str(i))
    w_data = w_data.append(data[0], ignore_index=True)
w_data
```

```python
#From course selecting last three number and assign them to new column C_code
a = w_data['Course'].str[3:]
w_data['C_code'] = a
w_data
```

```python
# Making a new column Degree and assigning value greater than 500 in C_code as Grad and otherwise Ugrad
w_data['Degree'] = ''
w_data['Degree'][w_data['C_code']<'500'] = 'Ugrad'
w_data['Degree'][w_data['C_code']>'500'] = 'Grad'
w_data
```

```python
# Total number of Grad and Ugrad
w_data['Degree'].value_counts()
```

```python
# Grad classes on tuesday
Tu_class = w_data[(w_data['Degree']=='Grad') & (w_data['Day'].str.find('Tu')>=0)]
Tu_class
```

```python
# Grad classes on tuesday starts at 12am
cls_twlv = Tu_class[Tu_class['Time'].str.startswith('12:00am')]
cls_twlv
```
